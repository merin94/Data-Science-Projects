## Domain Analysis

#### 1. Introduction
1.1 Background

The auto price prediction project aims to develop a machine learning model capable of predicting the prices of automobiles based on a set of features. This report provides a detailed analysis of the domain, focusing on the various features that influence auto prices.

1.2 Objectives

The primary objectives of the project include:
Understanding the significance of each feature in determining auto prices.
Selecting and engineering relevant features for model training.
Choosing an appropriate machine learning algorithm for price prediction.
Evaluating the model's performance using suitable metrics.

#### 2. Industry Overview
2.1 Market Trends

The automotive industry experiences dynamic trends influenced by consumer preferences, technological advancements, and economic factors. Current trends include a growing demand for fuel-efficient vehicles, increasing interest in electric vehicles, and the integration of advanced safety features.

2.2 Factors Influencing Auto Prices

Various factors contribute to the pricing of automobiles, including:
Brand reputation and perceived quality.
Vehicle specifications such as engine size, number of cylinders, and body style.
Fuel efficiency and type.
Technological features.
Market demand and supply.

#### 3. Data Considerations
3.1 Data Sources

The dataset for this project includes information on auto features and their corresponding prices.

3.2 Data Types

The dataset comprises a mix of continuous and categorical data. Features such as symboling, normalized-losses, and wheel-base are continuous, while make, fuel-type, and aspiration are categorical.

3.3 Data Collection Challenges

Potential challenges may include missing values in the normalized-losses column, outliers in numerical features, and the need for encoding categorical variables.


#### 4. Feature Engineering
4.1 Relevant Features

Key features influencing auto prices include make, fuel-type, engine-size, horsepower, city-mpg, highway-mpg, and others.

4.2 Feature Transformation

Feature transformation may involve normalizing continuous variables (wheel-base, length, width, etc.) and encoding categorical variables (e.g., one-hot encoding for make and label encoding for num-of-doors).

#### 5. Model Selection
5.1 Algorithm Choice

The choice of machine learning algorithm will be based on the regression nature of the problem. Linear regression, decision trees, or ensemble methods may be considered.

5.2 Model Evaluation Metrics

Evaluation metrics such as Mean Squared Error (MSE) and R-squared will be used to assess the model's accuracy in predicting auto prices.

#### 6. Business Implications
6.1 Use Cases

The auto price prediction model can be used by dealerships and consumers for informed decision-making during vehicle transactions. It can also assist manufacturers in setting competitive prices.

6.2 Stakeholders

Stakeholders include auto dealerships, manufacturers, and consumers. Dealerships can benefit from optimized pricing strategies, manufacturers from market insights, and consumers from fair pricing.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
error_rate = []
for K in range(0,20):
    K = K+1
    model = KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train)
    pred=model.predict(X_test)
    error = np.sqrt(mean_squared_error(y_test,pred))
    error_rate.append(error)
    print('RMSE value for k= ' ,K , 'is:', error)
     
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(0,20),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code

# CHALLENGES FACED DURING THIS PROJECT


#### HEADINGS: 
The dataset initially lacked clear headings, requiring a separate coding step after referring to the world file to assign appropriate headers to the columns, thereby enhancing data interpretability.

#### MISSING VALUES: 
The dataset contained missing values represented by '?'. To handle this, a preprocessing step was implemented by replacing these instances with NaN (Not a Number). Subsequently, missing values were imputed using mean and mode values to ensure a more complete dataset for analysis.

#### FEATURE SELECTION: 
Given that each feature independently impacted the target variable (price), the challenge arose in identifying specific features for model training. Despite the intricate relationships, the team explored correlations between features and dropped some based on their interdependence to streamline the dataset.

#### OUTLIERS: 
While the dataset included extreme values, resembling outliers, it was acknowledged that certain automotive models naturally exhibit higher or lower prices. To address this, an experiment was conducted by evaluating the model's accuracy with and without considering outliers. The optimal model performance was achieved when outliers were excluded, aligning with the understanding that price variations due to model-specific factors were not indicative of anomalous data points.
